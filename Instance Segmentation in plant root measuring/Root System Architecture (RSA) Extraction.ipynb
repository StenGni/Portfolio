```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize
from skan import Skeleton, summarize


def process_image(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    _, binary_image = cv2.threshold(image, 25, 255, cv2.THRESH_BINARY)
    
    # Skeletonize binary image
    skeleton_image = skeletonize(binary_image // 255)
    
    return skeleton_image


# Analyze skeleton and extract branch data
def analyze_skeleton(skeleton_image):
    skeleton = Skeleton(skeleton_image)
    branch_data = summarize(skeleton)
    print("Branch data extracted.")
    return branch_data


# Display the processed image
def display_image(image, title="Image"):
    plt.figure(figsize=(6, 6))
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()


# Main script
if __name__ == "__main__":
    input_folder = r"C:\Users\mstan\OneDrive\Dokumenty\GitHub\2024-25b-fai2-adsai-MonikaStangenberg231648\datalab_tasks\task6\model4_saved_bboxes"
    
    output_folder = r"C:\Users\mstan\OneDrive\Dokumenty\GitHub\2024-25b-fai2-adsai-MonikaStangenberg231648\datalab_tasks\task7\sceletons_2"
    os.makedirs(output_folder, exist_ok=True)

    # Get all image files in the input folder
    image_files = [os.path.join(input_folder, f) for f in os.listdir(input_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]
    print(f"Found {len(image_files)} image(s) in folder {input_folder}")

    # Process each image
    for idx, image_path in enumerate(image_files):
        print(f"\nProcessing image {idx + 1}/{len(image_files)}: {image_path}")
        
        try:
            # Process the image
            skeleton = process_image(image_path)
            
            # Display the skeletonized image
            display_image(skeleton, title=f"Skeletonized Image {idx + 1}")
            
            # Analyze skeleton and extract branch data
            branch_data = analyze_skeleton(skeleton)
            
            # Display branch data
            print(branch_data)
            
            # Save branch data to CSV
            original_filename = os.path.splitext(os.path.basename(image_path))[0]
            output_csv = os.path.join(output_folder, f"{original_filename}_branch_data.csv")
            branch_data.to_csv(output_csv, index=False)
```
