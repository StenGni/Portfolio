```python
import pandas as pd
from scipy.stats import chi2_contingency

transparency_mapping = {
    'Strongly disagree': 'Low Transparency Importance',
    'Somewhat disagree': 'Low Transparency Importance',
    'Neither agree nor disagree': 'Low Transparency Importance',  
    'Somewhat agree': 'High Transparency Importance',
    'Strongly agree': 'High Transparency Importance'
}

trust_mapping = {
    'Strongly disagree': 'Low Trust',
    'Somewhat disagree': 'Low Trust',
    'Neither agree nor disagree': 'Low Trust',  
    'Somewhat agree': 'High Trust',
    'Strongly agree': 'High Trust'
}

important_data['Transparency_Category'] = important_data[
    'Indicate the extent to which you agree or disagree with the following statements: - Transparency in business operations is important when deciding to trust an SME'
].map(transparency_mapping)

important_data['Trust_Category'] = important_data[
    'Indicate the extent to which you agree or disagree with the following statements: - I find the information given by SMEs to be trustworthy'
].map(trust_mapping)

```

```python
transparency_counts = important_data['Transparency_Category'].value_counts()
print("Unique value counts for Transparency_Category:")
print(transparency_counts)

trust_counts = important_data['Trust_Category'].value_counts()
print("\nUnique value counts for Trust_Category:")
print(trust_counts)
```

```python
contingency_table = pd.crosstab(important_data['Transparency_Category'], important_data['Trust_Category'])

print("Contingency Table:")
print(contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Test Results:")
print(f"Chi-Square statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.power import GofChisquarePower
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(
    important_data['Indicate the extent to which you agree or disagree with the following statements: - Transparency in business operations is important when deciding to trust an SME'],
    important_data['Indicate the extent to which you agree or disagree with the following statements: - I find the information given by SMEs to be trustworthy']
)

chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2_stat}")
print(f"p-value: {p_val}")
print(f"Degrees of freedom: {dof}")

def calculate_effect_size(chi2_stat, n):
    return np.sqrt(chi2_stat / n)

n = contingency_table.sum().sum()  
effect_size = calculate_effect_size(chi2_stat, n)

print(f"Effect Size (Cramer's V): {effect_size}")

alpha = 0.05  
power = 0.80  

power_analysis = GofChisquarePower()

sample_size_needed = power_analysis.solve_power(effect_size=effect_size, nobs=None, alpha=alpha, power=power)

print(f"Required sample size for power of 0.80: {sample_size_needed:.2f}")

actual_power = power_analysis.solve_power(effect_size=effect_size, nobs=n, alpha=alpha, power=None)

print(f"Actual power with current sample size: {actual_power:.2f}")
```

```python
from statsmodels.stats.proportion import proportions_ztest
from statsmodels.stats.power import zt_ind_solve_power

x1 = 8       # Number of males who prefer "Open Communication and Company Values"
n1 = 89      # Total number of males
x2 = 14      # Number of females who prefer "Open Communication and Company Values"
n2 = 107     # Total number of females

# Proportions for each group
p1 = x1 / n1  # Proportion of males who prefer the value
p2 = x2 / n2  # Proportion of females who prefer the value

# Pooled proportion
pooled_p = (x1 + x2) / (n1 + n2)

# Two-proportion Z-test
stat, p_value = proportions_ztest([x1, x2], [n1, n2])

# Effect size as the difference in proportions
effect_size = p2 - p1

# Power analysis parameters
alpha = 0.05  # Significance level
power = 0.8   # Desired power level

# Calculate required sample size per group to achieve the desired power
sample_size_needed = zt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='two-sided')

# Print results
print(f"Z-statistic: {stat:.2f}")
print(f"p-value: {p_value:.3f}")
print(f"Effect size (difference in proportions): {effect_size:.3f}")
print(f"Required sample size per group for 0.8 power: {sample_size_needed:.2f}")
```

```python
import pandas as pd
import scipy.stats as stats

# Mapping transparency ratings to numerical scores
transparency_score_mapping = {
    'Strongly disagree': 1,
    'Somewhat disagree': 2,
    'Neither agree nor disagree': 3,
    'Somewhat agree': 4,
    'Strongly agree': 5
}

# Mapping transparency ratings to categories of importance
transparency_category_mapping = {
    'Strongly disagree': 'Low Transparency Importance',
    'Somewhat disagree': 'Low Transparency Importance',
    'Neither agree nor disagree': 'Low Transparency Importance',
    'Somewhat agree': 'High Transparency Importance',
    'Strongly agree': 'High Transparency Importance'
}

# Converting text ratings to numerical scores
df['Transparency Score'] = df[
    'Indicate the extent to which you agree or disagree with the following statements: - Transparency in business operations is important when deciding to trust an SME'
].map(transparency_score_mapping)

# Applying category mapping
df['Transparency Importance'] = df[
    'Indicate the extent to which you agree or disagree with the following statements: - Transparency in business operations is important when deciding to trust an SME'
].map(transparency_category_mapping)

# Filtering data for only 'Female' and 'Male' values
df_filtered = df[df['How do you describe yourself? - Selected Choice'].isin(['Female', 'Male'])].copy()

def confidence_interval(data):
    mean = data.mean()
    std_err = stats.sem(data)  
    return stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)  

# Calculating confidence intervals for age groups and transparency importance levels
age_conf_intervals = {}

for age in df_filtered['How old are you?'].unique():
    for importance in ['Low Transparency Importance', 'High Transparency Importance']:
        scores = df_filtered[(df_filtered['How old are you?'] == age) & 
                             (df_filtered['Transparency Importance'] == importance)]['Transparency Score']
        
        if len(scores) > 1:
            ci = confidence_interval(scores)
            age_conf_intervals[(age, importance)] = ci

# Calculating confidence intervals for genders and transparency importance levels
gender_conf_intervals = {}

for gender in ['Female', 'Male']:
    for importance in ['Low Transparency Importance', 'High Transparency Importance']:
        scores = df_filtered[(df_filtered['How do you describe yourself? - Selected Choice'] == gender) & 
                             (df_filtered['Transparency Importance'] == importance)]['Transparency Score']
        
        if len(scores) > 1:
            ci = confidence_interval(scores)
            gender_conf_intervals[(gender, importance)] = ci

print("95% Confidence Intervals for Transparency Scores by Age Group and Importance Level:")
for (age, importance), ci in age_conf_intervals.items():
    print(f"Age {age}, {importance}: 95% CI = {ci}")

print("\n95% Confidence Intervals for Transparency Scores by Gender and Importance Level:")
for (gender, importance), ci in gender_conf_intervals.items():
    print(f"Gender {gender}, {importance}: 95% CI = {ci}")
```
